```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)

taxon_final = pd.read_csv('/content/NovabiomStat/final_bacteria_health.csv')
taxon140 = pd.read_csv('/content/NovabiomStat/final_bacteria_ibs_140.csv')
taxon141 = pd.read_csv('/content/NovabiomStat/final_bacteria_ibs_141.csv')

info_final = pd.read_csv('/content/NovabiomStat/final_health_statistic - final_health_statistic.csv.csv')
info140 = pd.read_csv('/content/NovabiomStat/final_ibs_140_statistic - final_ibs_140_statistic.csv.csv')
info141 = pd.read_csv('/content/NovabiomStat/final_ibs_141_statistic - final_ibs_141_statistic.csv.csv')
```

```python
male_final = list(info_final[info_final['Sex'] == 'male']['patient_ID'])
female_final = list(info_final[info_final['Sex'] == 'female']['patient_ID'])
mixed_final = list(info_final[info_final['Sex'] == 'mixed']['patient_ID'])

print('Male patients:', male_final)
print('Female patients:', female_final)
print('Mixed patients', mixed_final)
```

```python
#дроп нулей
taxon140 = taxon140.drop(taxon140.iloc[:, 1:].columns[taxon140.iloc[:, 1:].apply(lambda col: col.eq(0).all())], axis=1)
taxon141 = taxon141.drop(taxon141.iloc[:, 1:].columns[taxon141.iloc[:, 1:].apply(lambda col: col.eq(0).all())], axis=1)
taxon_final = taxon_final.drop(taxon_final.iloc[:, 1:].columns[taxon_final.iloc[:, 1:].apply(lambda col: col.eq(0).all())], axis=1)
```

```python
# @title без нулей мужчины и женщины
taxon140_male = taxon140[taxon140.iloc[:, 0].isin(male140)]
taxon140_female = taxon140[taxon140.iloc[:, 0].isin(female140)]
taxon140_male = taxon140_male.drop(taxon140_male.iloc[:, 1:].columns[taxon140_male.iloc[:, 1:].apply(lambda col: col.eq(0).all())], axis=1)
taxon140_female = taxon140_female.drop(taxon140_female.iloc[:, 1:].columns[taxon140_female.iloc[:, 1:].apply(lambda col: col.eq(0).all())], axis=1)

taxon141_male = taxon141[taxon141.iloc[:, 0].isin(male141)]
taxon141_female = taxon141[taxon141.iloc[:, 0].isin(female141)]
taxon141_male = taxon141_male.drop(taxon141_male.iloc[:, 1:].columns[taxon141_male.iloc[:, 1:].apply(lambda col: col.eq(0).all())], axis=1)
taxon141_female = taxon141_female.drop(taxon141_female.iloc[:, 1:].columns[taxon141_female.iloc[:, 1:].apply(lambda col: col.eq(0).all())], axis=1)

taxon_final_male = taxon_final[taxon_final.iloc[:, 0].isin(male_final)]
taxon_final_female = taxon_final[taxon_final.iloc[:, 0].isin(female_final)]
taxon_final_male = taxon_final_male.drop(taxon_final_male.iloc[:, 1:].columns[taxon_final_male.iloc[:, 1:].apply(lambda col: col.eq(0).all())], axis=1)
taxon_final_female = taxon_final_female.drop(taxon_final_female.iloc[:, 1:].columns[taxon_final_female.iloc[:, 1:].apply(lambda col: col.eq(0).all())], axis=1)
```

```python
# @title Определение нормальности/ненормальности, наличия/отсутсвия статистически значимых отличий мужчин
from scipy.stats import shapiro
from statsmodels.stats.multitest import multipletests
normal_list141 = []
not_normal_list141 = []

normal_list_final = []
not_normal_list_final = []

for col in taxon141_male.columns[1:]:
    stat, p = shapiro(taxon141_male[col])
    alpha = 0.05
    corrected_p_value = multipletests([p], method='bonferroni')[1][0]

    if corrected_p_value < 0.05:
      not_normal_list141.append(col)
    else:
      normal_list141.append(col)

for col in taxon_final_male.columns[1:]:
    stat, p = shapiro(taxon_final_male[col])
    corrected_p_value = multipletests([p], method='bonferroni')[1][0]

    if corrected_p_value < 0.05:
      not_normal_list_final.append(col)
    else:
      normal_list_final.append(col)
print(taxon141_male.shape, taxon_final_male.shape)
print('Нормальное распределение 141:', len(normal_list141))
print('Ненормальное распределение 141:', len(not_normal_list141))
print('Нормальное распределение final:', len(normal_list_final))
print('Ненормальное распределение final:', len(not_normal_list_final))

from scipy.stats import ttest_ind

# создаем два списка для результатов
no_dif141 = []
yes_dif141 = []
no_dif_final = []
yes_dif_final = []
# проводим t-тест для каждой пары столбцов
for i in range(len(normal_list141)):
    for j in range(i+1, len(normal_list141)):
        X = taxon141_male[normal_list141[i]]
        Y = taxon141_male[normal_list141[j]]
        t_statistic, p_value = ttest_ind(X, Y)
        if p_value < 0.05:
            yes_dif141.append([normal_list141[i], normal_list141[j]])
        else:
            no_dif141.append([normal_list141[i], normal_list141[j]])
print('Количество пар отличающихся 141:', len(yes_dif141))

for i in range(len(normal_list_final)):
    for j in range(i+1, len(normal_list_final)):
        X = taxon_final_male[normal_list_final[i]]
        Y = taxon_final_male[normal_list_final[j]]
        t_statistic, p_value = ttest_ind(X, Y)
        if p_value < 0.05:
            yes_dif_final.append([normal_list_final[i], normal_list_final[j]])
        else:
            no_dif_final.append([normal_list_final[i], normal_list_final[j]])
print('Количество пар отличающихся final:', len(yes_dif_final))
```
