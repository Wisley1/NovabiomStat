```python
# @title Определение нормальности/ненормальности, наличия/отсутсвия статистически значимых отличий мужчин
from scipy.stats import shapiro
from statsmodels.stats.multitest import multipletests
from scipy.stats import mannwhitneyu
normal_list141 = []
not_normal_list141 = []

normal_list_final = []
not_normal_list_final = []

for col in taxon141_male.columns[1:]:
    stat, p = shapiro(taxon141_male[col])
    alpha = 0.05
    corrected_p_value = multipletests([p], method='bonferroni')[1][0]

    if corrected_p_value < 0.05:
      not_normal_list141.append(col)
    else:
      normal_list141.append(col)

for col in taxon_final_male.columns[1:]:
    stat, p = shapiro(taxon_final_male[col])
    corrected_p_value = multipletests([p], method='bonferroni')[1][0]

    if corrected_p_value < 0.05:
      not_normal_list_final.append(col)
    else:
      normal_list_final.append(col)
print(taxon141_male.shape, taxon_final_male.shape)
print('Нормальное распределение 141:', len(normal_list141))
print('Ненормальное распределение 141:', len(not_normal_list141))
print('Нормальное распределение final:', len(normal_list_final))
print('Ненормальное распределение final:', len(not_normal_list_final))

from scipy.stats import ttest_ind

# создаем два списка нормального распределения результатов
no_dif141 = []
yes_dif141 = []
nno_dif141 = []
nyes_dif141 = []
nno_dif_final = []
nyes_dif_final = []
no_dif_final = []
yes_dif_final = []
# проводим t-тест для каждой пары столбцов
for i in range(len(normal_list141)):
    for j in range(i+1, len(normal_list141)):
        X = taxon141_male[normal_list141[i]]
        Y = taxon141_male[normal_list141[j]]
        t_statistic, p_value = ttest_ind(X, Y)
        if p_value < 0.05:
            yes_dif141.append([normal_list141[i], normal_list141[j]])
        else:
            no_dif141.append([normal_list141[i], normal_list141[j]])
print('Количество пар отличающихся 141 в нормальном распредлении:', len(yes_dif141))

for i in range(len(normal_list_final)):
    for j in range(i+1, len(normal_list_final)):
        X = taxon_final_male[normal_list_final[i]]
        Y = taxon_final_male[normal_list_final[j]]
        t_statistic, p_value = ttest_ind(X, Y)
        if p_value < 0.05:
            yes_dif_final.append([normal_list_final[i], normal_list_final[j]])
        else:
            no_dif_final.append([normal_list_final[i], normal_list_final[j]])
print('Количество пар отличающихся final в нормальном распределении:', len(yes_dif_final))


#for i in range(len(not_normal_list141)):
   # for j in range(i+1, len(not_normal_list141)):
     #   X = taxon141_male[not_normal_list141[i]]
      #  Y = taxon141_male[not_normal_list141[j]]
      #  stat, p_value = mannwhitneyu(X, Y)
      #  if p_value < 0.05:
       #     nyes_dif141.append([not_normal_list141[i], not_normal_list141[j]])
       # else:
      #      nno_dif141.append([not_normal_list141[i], not_normal_list141[j]])
#print('Количество пар отличающихся 141 в ненормальном распредлении:', len(nyes_dif141))
#for i in range(len(not_normal_list_final)):
    #for j in range(i+1, len(not_normal_list_final)):
     #   X = taxon_final_male[not_normal_list_final[i]]
      #  Y = taxon_final_male[not_normal_list_final[j]]
      #  stat, p_value = mannwhitneyu(X, Y)
      #  if p_value < 0.05:
       #     nyes_dif_final.append([not_normal_list_final[i], not_normal_list_final[j]])
       # else:
       #     nno_dif_final.append([not_normal_list_final[i], not_normal_list_final[j]])
#print('Количество пар отличающихся final в ненормальном распредлении:', len(nyes_dif_final))
```
